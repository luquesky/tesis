Vamos a hacer el análisis de la tabla de F0_MEAN... Carguemos la tabla primero

```python
%matplotlib inline

import pylab
import config
import logging
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

feature = "F0_MEAN"
```

Carguemos el dataframe de F0_MEAN


```python
df = pd.DataFrame.from_csv("tables/%s.csv" % feature)

df
```

Haremos lo siguiente:

- Primero agruparemos por sesión, calculando el promedio del entrainment y de cada variable social

```python
avg_by_session = df.groupby(by=["session", "speaker"]).mean()

avg_by_session
```

```python
import social

def correlation_analysis(df):
    corr = pd.Series(index=social.relevant_variables)

    for var in social.relevant_variables:
        corr[var] = df['entrainment'].corr(df[var])

    return corr

correlation_analysis(avg_by_session)
```

Ahora hagámoslo pero sobre el valor absoluto del entrainment...

```python
abs_corr = pd.Series(index=social.relevant_variables)

for var in social.relevant_variables:
    abs_corr[var] = abs(avg_by_session['entrainment']).corr(avg_by_session[var])

abs_corr
```

Hagamos una última cosa:
1. Primero, filtremos sólo los entrainments significativos, es decir, aquellas filas tal que $ entrainment > \frac{2}{\sqrt{n}}$

```python
from math import sqrt

significatives = df[abs(df.entrainment)**2 * df['count'] > 4]

corr = pd.Series(index=social.relevant_variables)

for var in social.relevant_variables:
    corr[var] = significatives['entrainment'].corr(significatives[var])

corr
```

```python
corr = pd.Series(index=social.relevant_variables)

for var in social.relevant_variables:
    corr[var] = abs(significatives['entrainment']).corr(significatives[var])

corr
```
