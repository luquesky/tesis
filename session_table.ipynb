Carguemos los datos sociales!

```python
social_variables = pd.DataFrame.from_csv("data/social-variables.csv")

display(social_variables)
```

Hagamos un join de ambas...

```python
df = pysqldf("""
    SELECT * from analysis INNER JOIN social_variables ON 
        analysis.session=social_variables.session AND 
        analysis.task=social_variables.task""")

display(df)
```

```python
sv = filter(lambda x: x!= "session" and x!="task", list(social_variables.columns))

correlations = pd.Series(index=sv)
df2 = df[df.speaker==0]

for social_var in sv:
    correlations[social_var] = df2['entrainment'].corr(df2[social_var])
    
correlations
```

```python
features = ["F0_MEAN", "F0_MAX", "F0_MAS", "F0_MEDIAN", 
            "ENG_MAX", "ENG_MIN", "ENG_MEAN", 
            "VCD2TOT_FRAMES", "SYLLABES_AVG", 'SYLLABES_COUNT', 
            'SOUND_ALL_LOCAL_JITTER', 'PHONEMES_AVG', 'PHONEMES_COUNT',
            'SOUND_VOICED_LOCAL_JITTER', 'NOISE_TO_HARMONICS_RATIO', 'SOUND_VOICED_LOCAL_SHIMMER',
            'SOUND_ALL_LOCAL_SHIMMER']
all_features_analysis = pd.DataFrame(columns=features, dtype=float)

for feature in features:
    analysis = Analyzer(feature).analyze(sessions)
    joined_analysis = pysqldf("""
    SELECT * from analysis INNER JOIN social_variables ON 
        analysis.session=social_variables.session AND 
        analysis.task=social_variables.task""")
    correlations = pd.Series(index=sv)
    df2 = joined_analysis[joined_analysis.speaker==0]

    for social_var in sv:
        correlations[social_var] = df2['entrainment'].corr(df2[social_var])
    
    all_features_analysis[feature] = correlations

display(all_features_analysis)
```
