```python
%matplotlib inline

import pylab
import config
import logging
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from tsa import entrainment, cross_correlogram
from session import SessionMapper, load_session
from tama import tama, normalize
from IPython.display import display,HTML, display_latex
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())


feature = "F0_MEAN"
logger = logging.getLogger('main')
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_columns', 100)


mapper = SessionMapper()

sessions = [load_session(i) for i in range(1,13)]
```

```python
from tsa import autoregressive_prewhitening


def generate_task_series(A, B, meanA, meanB, session, task):
    l_AB, E_AB, l_BA, E_BA = entrainment(A, B, lags=range(-6, 6))
    
    AB_data = pd.Series({
        "session": session.number,
        "task": task.number,
        "speaker": "A",
        "count": A.count(),
        "entrainment" : E_AB,
        "best_lag" : abs(l_AB),
        "tama_mean": meanA,
        "sample_mean": A.mean(),
    })
    
    BA_data = pd.Series({
        "session": session.number,
        "task": task.number,
        "speaker": "B",
        "count": B.count(),
        "entrainment" : E_BA,
        "best_lag" : abs(l_BA),
        "tama_mean": meanB,
        "sample_mean": B.mean()
    })
    
    return AB_data, BA_data

def analyze_task(session, task):
    task_data = pd.Series()

    A, meanA = tama(task.speechA, feature)
    B, meanB = tama(task.speechB, feature)

    if (A.count() < config.SERIES_LENGTH_THRESHOLD) or (B.count() < config.SERIES_LENGTH_THRESHOLD):
        raise StandardError("Series too short")

    nA = normalize(A, meanA)
    nB = normalize(B, meanB)

    rA, rB = autoregressive_prewhitening(nA, nB)
    task_key = task.name.replace("Task", "T").replace(" ", "")

    return generate_task_series(rA, rB, meanA=meanA, meanB=meanB, session=session, task=task)

def generate_table(sessions):
    session_table = pd.DataFrame(columns=["task","session", "count","speaker", "tama_mean", "sample_mean", "entrainment", "best_lag" ])
    
    for session in sessions:
        for task in session.tasks:
            try:
                entry_name = "S%s-T%s" % (session.number, task.number)
                session_table.loc[entry_name+"-A"], session_table.loc[entry_name+"-B"] = analyze_task(session, task)
            except StandardError as e:
                continue
    return session_table

analysis = generate_table(sessions)

display(analysis)
```
