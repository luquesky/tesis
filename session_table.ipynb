Saquemos primero las variables que terminan en '_no'

```python
import re
social_variables = pd.DataFrame.from_csv("data/social-variables.csv")
no_variables = filter(lambda x: re.match(r'.*_no', x), social_variables.columns)
yes_variables = filter(lambda x: re.match(r'.*_yes', x), social_variables.columns)

def rename_vars(social_variables):
    yes_rename = {}
    for v in yes_variables:
        yes_rename[v] = re.match(r'(?P<clean_var>.*)_yes', v).groupdict()['clean_var']
    
    social_variables.rename(columns=yes_rename, inplace=True)

social_variables.drop(no_variables, axis=1, inplace=True)
rename_vars(social_variables)

print(social_variables.columns)

display(social_variables)
```

Vamos a eliminar las variables con _no y quedarnos sólo con las que tienen _yes.

A su vez, convertiremos las variables de la forma $X_A$ y $X_B$ en $X$ a secas, de acuerdo al speaker en cuestión (recordar speaker=0 es A, y 1 es B)

La fórmula para convertir las variables es:

\begin{align}
X = (1 - speaker) X_A + speaker X_B
\end{align}

Si $speaker=0$, entonces se toma el $X_A$, y de la otra manera queda $X_B$

Primero, calculemos las variables de ese tipo (hay otras que no tienen _A ni _B!) que tienen que ser eliminadas, y las variables que resultan de ello (sin sufijo)

```python
social_vars = filter(lambda x: x!= "session" and x!="task", social_variables.columns)

print social_vars

vars_without_suffixes = set()
vars_to_be_removed = []
for var in social_vars:
    if re.match('.*_(A|B)', var):
        vars_without_suffixes.add(var[:-2])
        vars_to_be_removed.append(var)
    else:
        vars_without_suffixes.add(var)
print vars_without_suffixes
```

Ahora sí, convirtamos los dataframes usando la fórmula antes mencionada

```python
vars_without_suffixes = [
    "contributes_to_successful_completion",
    "making_self_clear",
    "engaged_in_game",
    "planning_what_to_say",
    "gives_encouragement",
    "difficult_for_partner_to_speak",
    "bored_with_game",
    "dislikes_partner",
]

def remove_A_B_variables(df, new_vars, old_vars):
    speaker = df['speaker']
    for sv in new_vars:
        try:
            sv_A = "%s_A" % sv
            sv_B = "%s_B" % sv
            df[sv] = (1 - speaker) * df[sv_A] + speaker * df[sv_B]
            assert((df[df.speaker == 0][sv] == df[df.speaker == 0][sv_A]).all())
            assert((df[df.speaker == 1][sv] == df[df.speaker == 1][sv_B]).all())
        except KeyError:
            print "%s is not of type _A or _B"
            pass
    df.drop(old_vars, axis=1, inplace=True)

remove_A_B_variables(df, new_vars=vars_without_suffixes, old_vars=vars_to_be_removed)


display(df)
```

```python
social_vars = vars_without_suffixes
features = ["F0_MEAN", "F0_MAX",
            "ENG_MAX", "ENG_MEAN", 
            "VCD2TOT_FRAMES", "SYLLABES_AVG", 'SYLLABES_COUNT', 
            'PHONEMES_AVG', 'PHONEMES_COUNT',
            'NOISE_TO_HARMONICS_RATIO', 'SOUND_VOICED_LOCAL_SHIMMER']
all_features_analysis = pd.DataFrame(columns=features, dtype=float)
twisted_all_features_analysis = pd.DataFrame(columns=features, dtype=float)

for feature in features:
    analysis = Analyzer(feature).analyze(sessions)
    joined_analysis = pysqldf("""
    SELECT * from analysis INNER JOIN social_variables ON 
        analysis.session=social_variables.session AND 
        analysis.task=social_variables.task""")
    twisted_joined_analysis = joined_analysis.copy()
    twisted_joined_analysis["speaker"] = 1 - twisted_joined_analysis["speaker"]

    remove_A_B_variables(joined_analysis, new_vars=social_vars, old_vars=vars_to_be_removed)
    remove_A_B_variables(twisted_joined_analysis, new_vars=social_vars, old_vars=vars_to_be_removed)
    
    joined_analysis.to_csv("tables/%s.csv" % feature)
    twisted_joined_analysis.to_csv("tables/twisted_%s.csv" % feature)
    
    correlations = pd.Series(index=social_vars)
    twisted_correlations = pd.Series(index=social_vars)
    
    def get_social_var_correlation(df, social_var):
        return df['entrainment'].corr(df[social_var])
    
    for social_var in social_vars:
        correlations[social_var] = get_social_var_correlation(joined_analysis, social_var)
        twisted_correlations[social_var] = get_social_var_correlation(twisted_joined_analysis, social_var)
    
    all_features_analysis[feature] = correlations
    twisted_all_features_analysis[feature] = twisted_correlations

print "Normal Analysis"
display(all_features_analysis)
print "Twisted Analysis"
display(twisted_all_features_analysis)
```
