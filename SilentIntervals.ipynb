```python
from speech import SpeechBuilder

speechA = SpeechBuilder("data/s10.objects.1.A.wav").speech 
speechB = SpeechBuilder("data/s10.objects.1.B.wav").speech
```

```python
silent_intervals = set([])
for intervalA in speechA.silent_intervals:
    for intervalB in speechB.silent_intervals:
        intersection = intervalB.intersect(intervalA.interval)
        if not intersection.is_EmptySet:
            silent_intervals.add(intersection)
```

```python
print silent_intervals
```

```python
max_interval = max(silent_intervals, key= lambda i: i.measure )
```
