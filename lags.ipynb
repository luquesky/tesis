```python



features = ['VCD2TOT_FRAMES', 'F0_MEAN', 'F0_MAX', 'ENG_MAX', 'ENG_MEAN', 
            "NOISE_TO_HARMONICS_RATIO", "SOUND_ALL_LOCAL_JITTER", "SOUND_ALL_LOCAL_SHIMMER",
            "SOUND_VOICED_LOCAL_JITTER", "SOUND_VOICED_LOCAL_SHIMMER", "SYLLABES_COUNT", "SYLLABES_AVG",
            "PHONEMES_COUNT", "PHONEMES_AVG"
            ]

tamas = session.analyze(features)
mapper.save(session)
```

```python
from helpers import cross_correlation

display_features = features = [ 'F0_MEAN', 'F0_MAX', 'ENG_MAX', 'ENG_MEAN', "SYLLABES_COUNT", "SYLLABES_AVG",
            "PHONEMES_COUNT", "PHONEMES_AVG"
            ]
lags = range(-20, 20)

import matplotlib.pyplot as plt


for feature in display_features:
    for index, task in enumerate(session.tasks):
        TA, averagesA = tamas[task][feature]["A"]
        TB, averagesB = tamas[task][feature]["B"]
        ccs = []
        for lag in lags:
            ccs.append(cross_correlation(averagesA, averagesB, lag))
        plt.title("Task %i - %s" % (index, feature) )
        plt.xlabel("Lag")
        plt.ylabel("Correlation")
        plt.scatter(lags, ccs, c="g")
        plt.show()

```
