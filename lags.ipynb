```python
from session import SessionMapper
import config
import math
from tama import hybrid_tama
from commands.plot import plot_tamas


mapper = SessionMapper()
session = mapper.fetch("data/s10.objects.1.tasks")

```

```python



features = ['VCD2TOT_FRAMES', 'F0_MEAN', 'F0_MAX', 'ENG_MAX', 'ENG_MEAN', 
            "NOISE_TO_HARMONICS_RATIO", "SOUND_ALL_LOCAL_JITTER", "SOUND_ALL_LOCAL_SHIMMER",
            "SOUND_VOICED_LOCAL_JITTER", "SOUND_VOICED_LOCAL_SHIMMER", "SYLLABES_COUNT", "SYLLABES_AVG",
            "PHONEMES_COUNT", "PHONEMES_AVG"
            ]

analysis = session.analyze(features)
mapper.save(session)
```

```python
from helpers import cross_correlation

display_features = features = [ 'F0_MEAN' ]

import matplotlib.pyplot as plt


for feature in display_features:
    for index, task in enumerate(session.tasks):
        task_analysis = analysis[task][feature]
        confidence_limit = task_analysis["confidence_value"]
        best_positive = task_analysis["best_positive_lag"]
        best_negative = task_analysis["best_negative_lag"]
        plt.title("Task %i - %s" % (index+1, feature) )
        plt.xlabel("Lag")
        
        plt.ylabel("Correlation")
        plt.axvline(0, color="k")
        plt.axhline(0, color="k")
        lags, ccs = zip(*task_analysis["cross_correlations"])
        plt.scatter(lags, ccs, c="b")
        l_m, l_M = min(lags), max(lags)
        plt.hlines(confidence_limit, l_m, l_M, linestyles='dashed', color='g')
        plt.hlines(-confidence_limit, l_m, l_M, linestyles='dashed', color='g')
    
        
        plt.scatter(*zip(best_positive, best_negative), color="r", marker="D")
        plt.show()

```
