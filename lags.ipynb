```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import config
from session import SessionMapper
from tama import tama, hybrid_tama
import pandas as pd


mapper = SessionMapper()

sessions = []

def load_session(session_number):
    num = str(session_number).zfill(2)
    name = "session-%s" % num
    tasks_path = "data/games-corpus/session_%s/s%s.objects.1.tasks" % (num, num)
    
    return mapper.fetch(tasks_path, name=name)

for i in range(1, 13):
    sessions.append(load_session(i))
```

```python
from helpers import cross_correlation
from operator import itemgetter

features = ["F0_MEAN", "ENG_MEAN"]

def analyze_task_feature(task, feature):
    print "Calculating %s" % feature

    A, A_mean = tama(task.speechA, feature)
    B, B_mean = tama(task.speechB, feature)

    assert np.array_equal(A.index, B.index)
    
    n = len(A)

    lag_limit = min(6, n)
    lags = range(-lag_limit, lag_limit+1)
    cross_correlations = [cross_correlation(A, B, lag) for lag in lags]

    # lag > 0
    zcc = zip(lags, cross_correlations)
    positive_correlations = [(l, cc) for (l, cc) in zcc if l >= 0]
    negative_correlations = [(l, cc) for (l, cc) in zcc if l <= 0]

    best_positive = max(positive_correlations, key=itemgetter(1))
    best_negative = max(negative_correlations, key=itemgetter(1))

    confidence_value = 2 / np.sqrt(n)
    
    plt.title("%s - %s" % (task.name, feature) )
    plt.xlabel("Lag")
    plt.ylabel("Correlation")
    plt.axvline(0, color="k")
    plt.axhline(0, color="k")
    plt.scatter(lags, cross_correlations, c="b")
    
    l_m, l_M = min(lags), max(lags)
    plt.hlines(confidence_value, l_m, l_M, linestyles='dashed', color='g')
    plt.hlines(-confidence_value, l_m, l_M, linestyles='dashed', color='g')


    plt.scatter(*zip(best_positive, best_negative), color="r", marker="D")
    plt.show()

    
    
for feature in features:
    print "\n" * 5
    print feature
    
    for session in sessions:
        print session.name
        for task in session.tasks:
            print "%s - %s - %s" % (feature, session.name, task.name)
            if task.length > config.TASK_LENGTH_THRESHOLD:
                analyze_task_feature(task, feature)
                

    
```
